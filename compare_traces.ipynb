```python
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_parquet(
    "hf://datasets/jonathanyin/aime_1983_2023_qwq-32b_fcs_traces/data/train-00000-of-00001.parquet"
)
df
```

```python
# Calculate the ratio between reasoning token count and first correct solution token count
df["Token Count Ratio"] = (
    df["First Correct Solution Token Count"] / df["Reasoning Token Count"]
)

# Display some statistics about the ratio
ratio_stats = df["Token Count Ratio"].describe()
print("Statistics for Token Count Ratio (First Correct Solution/Reasoning):")
print(ratio_stats)

# Display the dataframe with the new column
df
```

```python
# Plot the distribution of token count ratios
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df["Reasoning Token Count"], bins=30, kde=True)
plt.title("Distribution of Reasoning Token Count")
plt.xlabel("Reasoning Token Count")
plt.ylabel("Frequency")
plt.axvline(
    df["Reasoning Token Count"].mean(),
    color="r",
    linestyle="--",
    label=f"Mean: {df['Reasoning Token Count'].mean():.3f}",
)
plt.axvline(
    df["Reasoning Token Count"].median(),
    color="g",
    linestyle="--",
    label=f"Median: {df['Reasoning Token Count'].median():.3f}",
)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(df["First Correct Solution Token Count"], bins=30, kde=True)
plt.title("Distribution of First Correct Solution Token Count")
plt.xlabel("First Correct Solution Token Count")
plt.ylabel("Frequency")
plt.axvline(
    df["First Correct Solution Token Count"].mean(),
    color="r",
    linestyle="--",
    label=f"Mean: {df['First Correct Solution Token Count'].mean():.3f}",
)
plt.axvline(
    df["First Correct Solution Token Count"].median(),
    color="g",
    linestyle="--",
    label=f"Median: {df['First Correct Solution Token Count'].median():.3f}",
)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
